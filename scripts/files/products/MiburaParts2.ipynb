```python
df2 = df[['SonicWALL SKU', 'SonicWALL Product Description ', 'MSRP (USD)', 'Hardware/Software']]
```

```python
df3 = df2.loc[df2['Hardware/Software'] == 'H']
```

```python
categories = {}
```

```python
df3['category'] = 'Network'
```

```python
ar = []
for i,d in df3.iterrows():
    temp_str = d['SonicWALL Product Description ']
    if(d['SonicWALL Product Description '][:9] == 'SONICWALL'):
        temp_str = d['SonicWALL Product Description '][10:]
    
    if('WITH' in temp_str):
        index = temp_str.index('WITH')
        temp_str = temp_str[:index]
    
    if('SUPPORT' in temp_str):
        index = temp_str.index('SUPPORT')
        temp_str = temp_str[:index]
        
    if('PLUS' in temp_str):
        index = temp_str.index('PLUS')
        temp_str = temp_str[:index]
    
    if('UPGRADE' in temp_str):
        index = temp_str.index('UPGRADE')
        temp_str = temp_str[:index]
        
    if('YR' in temp_str):
        index = temp_str.index('YR')
        temp_str = temp_str[:index-2]
        
    if('(' in temp_str):
        index = temp_str.index('(')
        temp_str = temp_str[:index-1]
    
    ar.append(temp_str)
    
df3['SonicWALL Product Description '] = ar
```

```python
df3.head(3)
```

```python
df4 = df3.drop_duplicates(subset=['SonicWALL Product Description '])
```

```python
df4.shape
```

```python
df4.columns = ['brand', 'model', 'retail_price', 'partner_price', 'category']
```

```python
df4['brand'] = "SonicWall"
```

```python
df4['partner_price'] = 0
```

```python
df4['partner_price'] = df4['retail_price']
```
