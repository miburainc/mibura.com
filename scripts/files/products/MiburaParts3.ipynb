```python
import pandas as pd
import numpy as np

file = open("EPL_RAW.txt", 'r')
data = file.read()
file.close()
```

```python
data2 = data.split('\n')
```

```python
real_data = []

index1 = 0
index2 = 0

try:
    index1 = data2.index('(STANDARD) DISCOUNT CLASS FAMILY')+1
except:
    go = False

try:
    index2 = data2[10:].index('For U.S. Mississippi EPL Customer Use Only EMC Corporation') - 1
except:
    go = False

real_data.append(data2[index1:index2])  
data2 = data2[index2+22:]

go = True
while(go):
    
    try:
        index1 = data2.index('Net Price EPL Pricing')+1
    except:
        go = False
  
    try:
        index2 = data2[10:].index('For U.S. Mississippi EPL Customer Use Only EMC Corporation') - 1
    except:
        go = False

    real_data.append(data2[index1:index2])  
    data2 = data2[index2+22:]
    
```

```python
real_data = np.sum(real_data)
```

```python
real_data = real_data[:6435]
```

```python
rows = []
for i,string in enumerate(real_data):
    row = []
    s = string.split(' ')
    
    row.append(s[0])
    
    temp_string = ""
    temp_index = 1
    
    try:
        while("$" not in s[temp_index]):
            temp_string += s[temp_index] + " "
            temp_index += 1
    except:
        print(s)
        print(temp_index)
        print(i)
        
    row.append(temp_string.strip())
        
    while("$" in s[temp_index]) or ("%" in s[temp_index]):
        row.append(s[temp_index])
        temp_index += 1
    
    row.append(s[temp_index])
    temp_index += 1
    
    temp_string = ""
    
    while(temp_index < len(s)):
        temp_string += s[temp_index] + " "
        temp_index += 1
        
    row.append(temp_string.strip())
    
    rows.append(row)
    
print(rows[:20])
```

```python
df = pd.DataFrame(data=rows, columns=["model_code", "description", "retail_price", "discount", "partner_price","premium", "standard", "basic", "warr", "warrstand", "discountclass", "family"])
```
