```python
import pandas as pd
import numpy as np

data = pd.read_csv("microsoft_sep2017.csv")
```

```python
data.head()
```

```python
data.columns=data.loc[1]
```

```python
data.head()
```

```python
data = data[['Offer Display Name', 'End Customer Type']]
```

```python
data.head()
```

```python
data.columns = ['category', 'name']
```

```python
data['name'] = "Microsoft"
```

```python
data.fillna(value='', axis=1, inplace=True)
```

```python
data = data.loc[["(Government Pricing)" not in d for d in data['category']]]
```

```python
data = data.loc[["Gov" not in d for d in data['category']]]
```

```python
data = data.loc[1:]
```

```python
for index,row in data.iterrows():
    if 'Azure' in row['category']:
        data.loc[index,'name'] = "Microsoft Azure"
    elif 'Dynamics 365' in row['category'] or 'Dyn365' in row['category']:
        data.loc[index,'name'] = "Dynamics 365"
    elif 'Office 365' in row['category']:
        data.loc[index,'name'] = "Office 365"
    else:
        data.loc[index,'name'] = "other_microsoft"
```
